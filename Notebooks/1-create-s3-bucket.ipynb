# Create a session

Remember to add your aws access keys. Click on "Add-ons" above, then secrets, and add new secrets with label "access_key" and "secret_acces_key". 

Set the value as the corresponding keys.

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
```

```python
region = input("Type in your region: ")
```

```python
try:
    session = boto3.Session(
        aws_access_key_id=user_secrets.get_secret("access_key"),
        aws_secret_access_key=user_secrets.get_secret("secret_access_key"),
        region_name=region
    )
except Exception as e:
    print("Error occurred while creating session: ", e)
```

# Create S3 Bucket

```python
#AmazonPersonalizeFullAccess provides access to any S3 bucket with a name that includes "personalize" or "Personalize" 
bucket_name = input("Type in desired Bucket name: ")
```

```python
s3 = session.client('s3')
def create_s3_bucket(bucket_name, region=None):
    try:
        if region is None:
            s3.create_bucket(Bucket=bucket_name)
        else:
            s3.create_bucket(Bucket=bucket_name, CreateBucketConfiguration={"LocationConstraint": region})
        print(f"Bucket {bucket_name} created successfully.")
    except Exception as e:
        print(f"Error creating bucket: {e}")
```

```python
create_s3_bucket(bucket_name, region)
```
